```python
import pandas as pd #数据分析
import numpy as np #科学计算
from pandas import Series,DataFrame

df_train_transaction = pd.read_csv('./ieee-fraud-detection/train_transaction.csv')
df_train_identity = pd.read_csv('./ieee-fraud-detection/train_identity.csv')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
fig.set(alpha=0.2)

Survived_card1_0 = df_train_transaction.card4[df_train_transaction.isFraud == 0].value_counts()
Survived_card1_1 = df_train_transaction.card4[df_train_transaction.isFraud == 1].value_counts()
df = pd.DataFrame({u'虚假':Survived_card1_0,u'真实':Survived_card1_1})
df.plot(kind = 'bar',stacked=True)
plt.title(u'card1所代表的值对虚假信息的影响')
plt.xlabel(u'银行卡归属')
plt.ylabel(u'条目')
plt.show()
```
